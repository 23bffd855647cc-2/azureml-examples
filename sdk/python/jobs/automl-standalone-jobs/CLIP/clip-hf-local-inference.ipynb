```python
from PIL import Image
import requests
import time

from transformers import CLIPProcessor, CLIPModel

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

url1 = "./benchmark-images/can_small.jpg"
image1 = Image.open(url1)
url2 = "./benchmark-images/beach_large.jpg"
image2 = Image.open(url2)
url3 = "./benchmark-images/beach_huge.jpg"
image3 = Image.open(url3)

captions1 = "a photo of a can"
captions2 = "a photo of a beach"
captions3 = "a photo of a person"

images_list = [image1, image2, image3]
text_list = [captions1, captions2, captions3]
inputs = processor(text=text_list, images=images_list, return_tensors="pt", padding=True)

device = 'cuda:0'
model.to(device)
inputs.to(device)
time_sum = 0
# for i in range(100):
#     start_time = time.time()
#     outputs = model(**inputs)
#     time_sum += time.time() - start_time
# logits_per_image = outputs.logits_per_image # this is the image-text similarity score
# probs = logits_per_image.softmax(dim=1) # we can take the softmax to get the label probabilities
# import torch
# torch.set_printoptions(sci_mode=False)
# print(probs)

pytorch_total_params = sum(p.numel() for p in model.parameters())
print(pytorch_total_params)
print(model.device)
print(time_sum / 100)
```
